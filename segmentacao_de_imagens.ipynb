```python
from kMeansClustering import KMeansClustering
import matplotlib as mpl
from matplotlib import pyplot as plt
```

```python
def segment_image(image, n_clusters, max_iterations):
    # tamanho original da imagem
    old_shape = image.shape
    
    # adequando a imagem para o processo
    image = image.reshape(-1, 3)
    
    kmeans = KMeansClustering(n_clusters=n_clusters)
    # guardando as labels
    labels = kmeans.fit(image, max_iterations=max_iterations)
    # guardando a imagem segmentada
    segmented_img = kmeans.centroids[labels]
    # reformatando a imagem segmentada para os tamanhos da imagem original 
    segmented_img = segmented_img.reshape(old_shape)

    return segmented_img / 255
```

```python
img = mpl.image.imread("images/COMO_CUIDAR_DE_CACHORRO1.jpg")
# segmentando a imagem selecionada (demora)
segmented_images = [segment_image(img, 3, 10),
                    segment_image(img, 3, 10),
                    segment_image(img, 3, 10),
                    segment_image(img, 3, 10)]
# mostrando as imagens segmentadas
for i in range(len(segmented_images)):
    plt.subplot(2, 2, i + 1)
    plt.imshow(segmented_images[i])
```
